```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import shapiro
from scipy.stats import bartlett
from scipy.stats import ttest_ind
from scipy.stats import mannwhitneyu
from scipy.stats import kruskal
from scipy.stats import f_oneway
from statsmodels.stats.multitest import multipletests


from numpy.random import seed
from numpy.random import randn

q95 = [0.97, 0.829, 0.71, 0.625, 0.568, 0.526, 0.493, 0.466,
       0.444, 0.426, 0.41, 0.396, 0.384, 0.374, 0.365, 0.356,
       0.349, 0.342, 0.337, 0.331, 0.326, 0.321, 0.317, 0.312,
       0.308, 0.305, 0.301, 0.29
      ]
Q95 = {n:q for n,q in zip(range(3,len(q95)+1), q95)}
def dixon_test(data, left=True, right=True, q_dict=Q95):
    """
    Keyword arguments:
        data = A ordered or unordered list of data points (int or float).
        left = Q-test of minimum value in the ordered list if True.
        right = Q-test of maximum value in the ordered list if True.
        q_dict = A dictionary of Q-values for a given confidence level,
            where the dict. keys are sample sizes N, and the associated values
            are the corresponding critical Q values. E.g.,
            {3: 0.97, 4: 0.829, 5: 0.71, 6: 0.625, ...}

    Returns a list of 2 values for the outliers, or None.
    E.g.,
       for [1,1,1] -> [None, None]
       for [5,1,1] -> [None, 5]
       for [5,1,5] -> [1, None]

    """
    assert(left or right), 'At least one of the variables, `left` or `right`, must be True.'
    assert(len(data) >= 3), 'At least 3 data points are required'
    assert(len(data) <= max(q_dict.keys())), 'Sample size too large'

    sdata = sorted(data)
    Q_mindiff, Q_maxdiff = (0,0), (0,0)

    if left:
        Q_min = (sdata[1] - sdata[0])
        try:
            Q_min /= (sdata[-1] - sdata[0])
        except ZeroDivisionError:
            pass
        Q_mindiff = (Q_min - q_dict[len(data)], sdata[0])

    if right:
        Q_max = abs((sdata[-2] - sdata[-1]))
        try:
            Q_max /= abs((sdata[0] - sdata[-1]))
        except ZeroDivisionError:
            pass
        Q_maxdiff = (Q_max - q_dict[len(data)], sdata[-1])

    if not Q_mindiff[0] > 0 and not Q_maxdiff[0] > 0:
        outliers = [None, None]

    elif Q_mindiff[0] == Q_maxdiff[0]:
        outliers = [Q_mindiff[1], Q_maxdiff[1]]

    elif Q_mindiff[0] > Q_maxdiff[0]:
        outliers = [Q_mindiff[1], None]

    else:
        outliers = [None, Q_maxdiff[1]]

    return outliers
```

```python
def histo(data, title, title_y, title_x) :
    fig=plt.figure()
    ax=fig.add_axes([0,0,1,1])
    ax.hist(data)
    ax.set_title(title)
    ax.set_ylabel(title_y)
    ax.set_xlabel(title_x)
    
histo((jeune_old_rotarod), "Pourcentage du temps sur le rotarod pour les jeunes et agé", "nombre d'individu", "Pourcentage du temps")

```

```python

def graph_q1(jeune,age,jeune_e,age_e,titre,titre_y, percent_height = 0.8) :
    labels = ["Age"]


    x = np.arange(len(labels))  # the label locations
    width = 0.25  # the width of the bars

    fig, ax = plt.subplots()
    rects1 = ax.bar(x - width/2, jeune, width, label='Jeune', yerr = jeune_e)
    rects2 = ax.bar(x + width/2, age, width, label='Ages', yerr = age_e)

    # Add some text for labels, title and custom x-axis tick labels, etc.
    ax.set_ylabel(titre_y)
    ax.set_title(titre)
    ax.set_xticks(x)
    ax.set_xticklabels(labels)
    ax.legend()
    


    def autolabel(rects):
        """Attach a text label above each bar in *rects*, displaying its height."""
        for rect in rects:
            height = rect.get_height()
            ax.annotate('{}'.format(height),
                        xy=(rect.get_x() + rect.get_width() / 2, percent_height*height),
                        xytext=(0, 3),  # 3 points vertical offset
                        textcoords="offset points",
                        ha='center', va='bottom')


    autolabel(rects1)
    autolabel(rects2)

    fig.tight_layout()

    plt.show()

graph_q1(rotarod_young,rotarod_old,rotarod_young_e ,rotarod_old_e ,'Moyenne en % de temps sur le rotarod en fonction de l\'âge','Moyenne en % de temps sur le rotarod')
```

** : significativité avec p-value <= 0.01

*** : significativité avec p-value <= 0.001

```python
def moy_var(*stat):
    
    print("la moyenne des jeunes est : " + str(stat[0]))
    print("l'écart-type des jeunes est : " + str(stat[1]))
    print("la moyenne des agés est : " + str(stat[2]))
    print("l'écart-type des agés est : " + str(stat[3]))
    
    if len(stat)>4:
        print("la moyenne des restrictions caloriques est : " + str(stat[4]))
        print("l'écart-type des restrictions caloriques est : " + str(stat[5]))
        print("la moyenne des fat diet est : " + str(stat[6]))
        print("l'écart-type des fat diet est : " + str(stat[7]))
            
def gausienne(p_value,alpha = 0.05):
    if p_value > alpha :
        print("Comme " + str(p_value) + " > " + str(alpha) + " d'après le shapiro test, la distribution suit une gaussienne")
    else :
        print("Comme " + str(p_value) + " < " + str(alpha) + " d'après le shapiro test, la distribution ne suit pas une gaussienne")
def egal_var(p_value,alpha = 0.05):
    if p_value > alpha :
        print("Comme " + str(p_value) + " > " + str(alpha) + " d'après le test de Bartlett, il y a égalité des variances")
    else :
        print("Comme " + str(p_value) + " < " + str(alpha) + " d'après le test de Bartlett, il y n'y a pas égalité des variances")

def significativite(test, p_value, alpha = 0.05, alpha2 = 0.01, alpha3 = 0.001) :
    if p_value < alpha3 : 
        print("Comme " + str(p_value) + " < " + str(alpha3) + " d'après le test de " + test + " il y a significativité avec ***")
    elif p_value < alpha2 : 
        print("Comme " + str(p_value) + " < " + str(alpha2) + " d'après le test de " + test + " il y a significativité avec **")
    elif p_value < alpha : 
        print("Comme " + str(p_value) + " < " + str(alpha) + " d'après le test de " + test + " il y a significativité avec *")
    else :
        print("Comme " + str(p_value) + " > " + str(alpha) + " d'après le test de " + test + " il y a pas de significativité")

    
moy_var(rotarod_young,rotarod_young_e,rotarod_old,rotarod_old_e)
gausienne(0.59)
egal_var(round(bartlett(jeune_rotarod, old_rotarod)[1],3))
print("2 échantillons, distributions gaussiennes et égalité des variance : test de student (ech indpt) !")
significativite("Student",ttest_ind(jeune_rotarod, old_rotarod)[1])
```
